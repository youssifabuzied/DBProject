# Code for getting the phone numbers from the Description

```python
import re
```

```python
onames= []
onums = []
oexatr = []
for i in range(0, len(Ads)):
    txt = Ads["Description"][i]
    DPN = re.findall(r'(\+?2?0?\d{10})', txt)
   
    for i in range(0, len(DPN)):
        onames.append(Ads["Owner_Name"][i])
        onums.append(Ads["Owner_Phone_Number"][i])
        oexatr.append(DPN[i])
```

```python
OwnersExtraNumbers = pd.DataFrame()
OwnersExtraNumbers["Owner_Name"] = onames
OwnersExtraNumbers["Primary_Phone_Number"] = onums
OwnersExtraNumbers["Extra_Number"] = oexatr
```

```python
OwnersExtraNumbers.drop_duplicates(inplace= True)
```

```python
OwnersExtraNumbers.reset_index(inplace = True, drop = True)
OwnersExtraNumbers.to_csv("OwnersExtraNumbers.csv")
```

# Code for faking the user tables

```python
import random
```

```python
user_name= []
user_email = []
gender= []
birthdate= []
gender=[]
months= ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct","Nov", "Dec"]
chars= ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
names= ["Ahmed", "Mohamed", "Omar", "youssif", "Salah", "Salma", "Noura", "Mariam", "Rana", "Fatma"]
genders = ['m','f']
for i in range(0, 100):
    s = random.randint(0,1)
    gender.append(genders[s])
    if(s == 0):
        user_name.append(names[random.randint(0,4)]+ "_"+ names[random.randint(0,4)])
    else:
        user_name.append(names[random.randint(5,9)]+ "_"+ names[random.randint(0,4)])
        
    user_email.append(chars[random.randint(0,25)]+str(i)+"@gmail.com")
    day = random.randint(1,28)
    Year = random.randint(1960,2005)
    dates= str(day)+" "+months[random.randint(0,11)]+ " "+str(Year)
    
    birthdate.append((parse(dates, ignoretz=True)).strftime("%d-%m-%Y"))
```

```python
users = pd.DataFrame()
users["User_Name"] = user_name
users["Email_Address"] = user_email
users["Birthdate"] = birthdate
users["Gender"] = gender
```

```python
users
```

```python
users.to_csv("Final_Users.csv")
```

```python
makes= ["Mercedes-Benz", "BMW", "Toyota", "Nissan", "Kia"]
models= {"Mercedes-Benz" : ["A Class", "B Class", "C Class", "E Class", "S Class"],
        "BMW":["X7", "X6", "X6 M50i", "X5", "X4"], 
        "Toyota" : ["Corolla", "Yaris", "Camry", "Sienna", "Prius"], 
        "Nissan" : ["Sunny", "Qashqai", "Juke", "Sentra", "Micra"], 
        "Kia":["Cerato", "Sportage", "Rio", "Carnival", "Seltos"]}
user= []
make= []
model = []
for i in range(0, len(users)):
    s = random.randint(0,5)
    if(s== 0):
        user.append(users["Email_Address"][i])
        g = random.randint(0,4)
        make.append(makes[g])
        model.append("All")
    else:
        for j in range(0,s):
            user.append(users["Email_Address"][i])
            g = random.randint(0,4)
            make.append(makes[g]) 
            model.append(models[makes[g]][random.randint(0,4)])
```

```python
user_preferences = pd.DataFrame()
user_preferences["Email_Address"] = user
user_preferences["Make"] = make
user_preferences["Model"] = model
```
