```python
from selenium.common.exceptions import NoSuchElementException
from datetime import datetime, timedelta
for i in range(3758, len(links)): 
  
    #response = requests.get(links[i])
    try:
        current_date = datetime.now()
        driver.get(links[i])
        driver.implicitly_wait(8)
        posting_date = driver.find_element(By.CSS_SELECTOR, "#body-wrapper > div > header:nth-child(3) > div > div > div > div._0a9bc591 > div._408759e3 > div:nth-child(1) > div > div._1075545d.e3cecb8b._5f872d11 > span:nth-child(2) > span").text
        #print(posting_date)
        posting_date = posting_date.split(" ")
        listed_date = datetime.now()
        if( "months" in posting_date or "month" in posting_date or "year" in posting_date or "years" in posting_date):
            continue
        else:
            if("days" in posting_date):
                listed_date = current_date - timedelta(days=int(posting_date[0]))
            elif("day" in posting_date):
                listed_date = current_date - timedelta(days=1)
            elif("weeks" in posting_date):
                listed_date = current_date - timedelta(weeks=int(posting_date[0]))
            elif("week" in posting_date):
                listed_date = current_date - timedelta(weeks=1)
            elif("hours" in posting_date):
                listed_date = current_date - timedelta(hours=int(posting_date[0]))
            else:
                listed_date = current_date
        driver.find_element(By.CSS_SELECTOR, "#body-wrapper > div > header:nth-child(3) > div > div > div > div._0a9bc591 > div._408759e3 > div:nth-child(2) > div > div._1075545d.b34f9439._42f36e3b._96d4439a._1709dcb4 > span._09eb0c84._79855a31").click()
        time.sleep(1)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        owner_link =  soup.find_all(class_ = "_1075545d d059c029")
        
        if(len(owner_link) == 0):
            continue
        for key in AD_data.keys():
            AD_data[key].append(None)
        for key in Cars_data.keys():
            Cars_data[key].append(None)
        for key in Owners.keys():
            Owners[key].append(None)
            
        AD_data["posting date"][count] = listed_date.date()
        #print(posting_date)
        allele = soup.find_all(class_= "b44ca0b3")
       
        desc = soup.find_all(class_= "_0f86855a")
        loc = soup.find_all(class_= "fccf2e37")
        #print(loc)
        #print(desc)
        ai = soup.find_all(class_ = "_171225da")
        #phone_Number = soup.find_all(class_ ="_45d98091 _2e82a662" )
        #print(phone_Number)

        #print(len(loc))
        AD_data["Description"][count] = desc[0].next_element.next_element
        AD_data["Location"][count] = loc[0].next_element
        AD_data["AD_ID"][count] = ai[0].next_element
        AD_data["OwnerProfileLink"][count] = "https://www.olx.com.eg"+owner_link[0].find("a").get("href")
        AD_data["phone_number"][count] = str(driver.find_element(By.CSS_SELECTOR, "#body-wrapper > div > header:nth-child(3) > div > div > div > div._0a9bc591 > div._408759e3 > div:nth-child(2) > div > div._1075545d.b34f9439._42f36e3b._96d4439a._1709dcb4 > span").text)
        Owners["Phone Number"][count] = AD_data["phone_number"][count]
        seller_name = soup.find_all(class_ = "_1075545d _6caa7349 _42f36e3b d059c029")
        join_data = soup.find_all(class_ = "_34a7409b")
        Owners["Owner Name"][count] = seller_name[0].next_element.next_element
        Owners["Profile Link"][count] = AD_data["OwnerProfileLink"][count]
        for t in join_data:
            ele = t.next_element
            words = ele.split(" ")
            if(words[0] == "Member"):
                Owners["Join Date"][count] = ele
        price = soup.find_all(class_ = "_56dab877")
        features = soup.find_all(class_ = "_66b85548")
        for ele in allele:

            if(ele.next_element.next_element in AD_data.keys()):
                AD_data[ele.next_element.next_element][count] = ele.next_element.next_element.next_element.next_element
            if(ele.next_element.next_element in Cars_data.keys()):
                Cars_data[ele.next_element.next_element][count]= ele.next_element.next_element.next_element.next_element
        lis = []
        for f in features:
            lis.append(f.next_element)
        Cars_data["Features"][count] =  lis
        
        AD_data["link_number"][count] = i
        count+=1
        print(i)
    except NoSuchElementException:
        continue
#driver.quit()
```
